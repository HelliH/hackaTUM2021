### Forecasting solution: ARIMA models

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA

# set microscope and sensors names
microscope = 'MICDEV01'
sensor = 'LSM_HS_SensorCan81_Temperature_Room'
name = microscope+', '+sensor

t = data.loc[(data.source_id == microscope) & (data.sensor_name == sensor)]
t['datetime'] = pd.to_datetime(t['datetime'])

# set dates
date1 = pd.to_datetime('2021-07-01', format='%Y-%m-%d', utc=True)
date2 = pd.to_datetime('2021-09-30', format='%Y-%m-%d', utc=True)

t = t.loc[(t['datetime']>date1) & (t['datetime'] < date2)]
t.rename(columns = {'sensor_value': name, 'datetime': ''}, inplace = True)
t = t.sort_values('')
t.index = t['']

# numer of hours to predict
horizont = 1
yy = pd.DataFrame(index=[])
for k in range(20, len(t[name])):
    try:
        model = ARIMA(t[name][:k], order=(2, 0, 3))
        model_fit = model.fit()
        yhat_f = pd.DataFrame(model_fit.forecast(steps=horizont)[0])
        yhat_f.index = [t[name].index[k] + pd.DateOffset(hours=l) for l in range(horizont)]
        yy = pd.concat([yy, yhat_f])
    except:
        pass
yy = yy.sort_index()
plt.figure(figsize=(12,6))
plt.plot(yy, color='red', label='forecast')
plt.plot(t[name], color='blue', label='historical')
plt.legend()
plt.show()
```

For regression-like setting, we need datetime to be converted into numeric format

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
ts1 = scaler.fit_transform(np.array(t['ts']).reshape(-1,1))
```

### Gaussian processes are powerful tools to model data to fill missing values or detect anomalies

```python
from sklearn.gaussian_process import GaussianProcessRegressor

from sklearn.gaussian_process.kernels import DotProduct, WhiteKernel, RBF
scaler = MinMaxScaler()
ts1 = scaler.fit_transform(np.array(t['ts']).reshape(-1,1))
kernel = DotProduct() + WhiteKernel() + RBF(length_scale=1)
gpr = GaussianProcessRegressor(kernel=kernel, random_state=0, alpha=0.01)
gpr.fit(ts1, t[microscope+  ', LSM_HS_SensorCan81_Temperature_Room'])
```

```python
import matplotlib
# We make prediction for uniformly distributed datetimes inside the given interval
# Still, this data was not used during the training, so the predictions are out-of-sample
predictions, sd = gpr.predict(np.arange(0, 1, 0.001).reshape(-1, 1), return_std=True)
upper = [predictions[k]+1.96*sd[k] for k in range(len(predictions))]
lower = [predictions[k]-1.96*sd[k] for k in range(len(predictions))]
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 16}

matplotlib.rc('font', **font)
plt.figure(figsize=(12, 6))
plt.xlabel('TIME')
plt.title('ROOM TEMPERATURE SENSOR', fontweight="bold")
plt.ylabel('TEMPERATURE')
plt.plot(t.index, t[microscope+', LSM_HS_SensorCan81_Temperature_Room'], linewidth=3.5,color='k', label='Sensor data')
plt.plot([min(t.index)+k/len(predictions)*(max(t.index)-min(t.index)) for k in range(len(predictions))], predictions,linewidth=4.5,color='dodgerblue', label='Prediction')
plt.fill_between([min(t.index)+k/len(predictions)*(max(t.index)-min(t.index)) for k in range(len(predictions))], lower, upper, alpha=0.2, color='dodgerblue', label='Confidence interval')
plt.legend()
plt.savefig('No_anomalies.png')
plt.show()
```

### Evaluation of our missing values predictor

We randomly deleted ground truth data and tried to recover its values with our missing values predictor (ground truth point was removed during training). 93.4% of true points were correctly detected inside 95% predicted confidence interval

```python
scaler = MinMaxScaler()
ts1 = scaler.fit_transform(np.array(t['ts']).reshape(-1,1))
kernel = DotProduct() + WhiteKernel() + RBF(length_scale=1)
table = []
for i in range(1, len(ts1)-1):
    x_true = ts1[i]
    y_true = t[microscope+  ', LSM_HS_SensorCan81_Temperature_Room'].values[i]
    x = np.delete(ts1, i)
    y = np.delete(t[microscope+  ', LSM_HS_SensorCan81_Temperature_Room'].values, i)
    gpr2 = GaussianProcessRegressor(kernel=kernel, random_state=0, alpha=0.01)
    gpr2.fit(x.reshape(-1,1), y)
    pred, sd = gpr2.predict([x_true], return_std=True)
    table.append((y_true, pred, sd))
```

```python
from scipy.stats import percentileofscore
per = []
for y, m, sd in table:
    dat = np.random.normal(m, sd, 10**4)
    per.append(percentileofscore(dat, y))
```

```python
print("Model accuracy: ", len([k for k in per if (k<97.5) and (k > 2.5)])/len(per))
```
